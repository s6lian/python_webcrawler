```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://www.century21.ca/search/43.55305912986279;-80.67119881093868;43.41131507917129;-80.42228981435665/price~ASC/v_Gallery")
c = r.content
```

```python
soup=BeautifulSoup(c,"html.parser")
# print(soup)
```

```python
all= soup.find_all("div",{"class":"property-list-item"})
```

```python
len(all)
all[0].find("span",{"class":"price"}).text
```

```python
l=[]
for item in all:
    d = {}
    try:
        d["Property_ID"] = item.find("span",{"class":"property-id"}).text.replace("Property ID:","").replace(" ","")
    except:
        d["Property_ID"] = None
    try:
        d["Address"] = (item.find("span",{"class":"address-line-1"}).text)
        d["Locality"] = (item.find("span",{"class":"address-line-2"}).text)
        d["Price"] = (item.find("span",{"class":"price"}).text)
    except:
        pass
        
    try:
        d["beds"]=(item.find("li",{"class":"beds"}).text)
    except:
        d["beds"]= None
    try:
        
        d["baths"]=(item.find("li",{"class":"baths"}).text)
        
    except:
        d["baths"]= None
    try:
        d["sqft"]= (item.find("li",{"class":"sqft"}).text)
    except:
        d["sqft"]= None
        
    l.append(d) 
```

```python
for i in l:
    print(i)
```

```python
import pandas as pd
df = pd.DataFrame(l)
df.index = df['Locality']
df
```
